```python
polarity = []
flag = 0
for comment in comments['comment_text']:
    try:
        polarity.append(TextBlob(comment).sentiment.polarity)
        flag = 1
    except:
        polarity.append(0)
print(flag)
comments['polarity'] = polarity
```

```python
filter_positive = comments['polarity'] == 1
```

```python
comments_postive = comments[filter_positive]
```

```python
filter_negative = comments['polarity'] == -1
```

```python
comments_negative = comments[filter_negative]
```

```python
type(comments['comment_text'])
```

```python
total_comments_positive = ''.join(comments_postive['comment_text'])
total_comments_negative = ''.join(comments_negative['comment_text'])
```

```python
wordcloud_positive = WordCloud(stopwords = set(STOPWORDS)).generate(total_comments_positive)
wordcloud_negative = WordCloud(stopwords = set(STOPWORDS)).generate(total_comments_negative)
```

```python
plt.imshow(wordcloud_positive)
plt.axis('off')
```

```python
plt.imshow(wordcloud_negative)
plt.axis('off')
```

```python
emoji_list = [char for char in comments['comment_text'] if char in emoji.EMOJI_DATA]
```

```python
emoji_list[0:10]
```

```python
emojis = [Counter(emoji_list).most_common(10)[i][0] for i in range(10)]
print(emojis)
type(emojis)
```

```python
emoji_count = [Counter(emoji_list).most_common(10)[i][1] for i in range(10)]
print(emoji_count)
type(emoji_count)
```

```python
plt.bar(emojis, emoji_count)  #using matplotlib
plt.show()
```

```python
full_df = pd.DataFrame()
path = r'/Users/alka/Projects/Data-Analysis/additional_data'


for file in files_csv:
    current_df = pd.read_csv(path+'/'+file , encoding='iso-8859-1' , on_bad_lines = 'skip')
    
    full_df = pd.concat([full_df , current_df] , ignore_index=True)
```

```python
cat_dict = {}

for item in json_df['items'].values:
    ## cat_dict[key] = value (Syntax to insert key:value in dictionary)
    cat_dict[int(item['id'])] = item['snippet']['title']
```

```python
cat_dict
```

```python
full_df['category_name'] = full_df['category_id'].map(cat_dict)
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='category_name' , y='likes' , data=full_df)
plt.xticks(rotation='vertical')
```

```python
full_df['like_rate'] = (full_df['likes']/full_df['views'])*100
full_df['dislike_rate'] = (full_df['dislikes']/full_df['views'])*100
full_df['comment_count_rate'] = (full_df['comment_count']/full_df['views'])*100
```
